```python
import rasterio
from rasterio.windows import Window

def clip_center_of_raster(input_raster_path, output_raster_path, tile_size=2000):
    with rasterio.open(input_raster_path) as src:
        # Calculate the coordinates of the center of the raster
        center_x, center_y = src.width // 2, src.height // 2

        # Calculate the window's position to clip around the center
        window = Window(
            col_off=center_x - tile_size // 2,
            row_off=center_y - tile_size // 2,
            width=tile_size,
            height=tile_size
        )

        # Read the data from the window
        data = src.read(window=window)

        # Define a new transform for the new raster
        new_transform = src.window_transform(window)

        # Define metadata for the new raster
        new_meta = src.meta.copy()
        new_meta.update({
            "driver": "GTiff",
            "height": tile_size,
            "width": tile_size,
            "transform": new_transform
        })

        # Write the clipped data to a new file
        with rasterio.open(output_raster_path, 'w', **new_meta) as dst:
            dst.write(data)

# Usage
input_raster_path = 'reprojected_dem_4490.tif'
output_raster_path = 'clipped_dem_ningxia_small.tif'
clip_center_of_raster(input_raster_path, output_raster_path)

# Visualize the Clipped DEM
with rasterio.open('clipped_dem_ningxia_small.tif') as clipped_raster:
    clipped_data = clipped_raster.read(1, masked=True)  # Reading with mask applied to nodata
    plt.figure(figsize=(8, 8))
    plt.imshow(clipped_data, cmap='terrain')
    plt.colorbar(label='Elevation')
    plt.title('Clipped DEM of Ningxia (small)')
    plt.show()

import time
time.sleep(1)
```

```python
import geopandas as gpd
import rasterio
from rasterio.features import rasterize
import numpy as np
import torch

# Load the raster data for height
height_tif = rasterio.open("normalized_clipped_dem_ningxia_small.tif")
height = height_tif.height
width = height_tif.width
print(f"height: {height}")
print(f"width: {width}")

# Function to rasterize shapefiles based on the reference raster dimensions
def rasterize_shapefile(shapefile, reference_raster):
    transform = reference_raster.transform
    out_shape = (reference_raster.height, reference_raster.width)
    rasterized = rasterize([(geom, 1) for geom in shapefile.geometry],
                           out_shape=out_shape,
                           fill=0, transform=transform,
                           dtype='uint8')
    return rasterized

# Rasterize rivers, residents, and roads
rivers_raster = rasterize_shapefile(ningxia_rivers, height_tif)
residents_raster = rasterize_shapefile(ningxia_residents, height_tif)
roads_raster = rasterize_shapefile(ningxia_roads, height_tif)

# Combine input layers
input_layers = np.stack([rivers_raster, residents_raster, height_tif.read(1)], axis=-1)

# Flatten the arrays for splitting
# input_flat = input_layers.reshape(-1, input_layers.shape[2])
# output_flat = roads_raster.flatten()
input_flat = input_layers
output_flat = roads_raster
print(input_flat.shape)

# Define split sizes and shuffle indices
total_pixels = input_flat.shape[0]
train_size = int(total_pixels * 0.6)
val_size = int(total_pixels * 0.2)

indices = np.arange(total_pixels)
# np.random.shuffle(indices)

# Split the data
train_indices = indices[:train_size]
val_indices = indices[train_size:train_size + val_size]
test_indices = indices[train_size + val_size:]

train_x, train_y = input_flat[train_indices], output_flat[train_indices]
print(train_x.shape)
print(train_y.shape)
val_x, val_y = input_flat[val_indices], output_flat[val_indices]
test_x, test_y = input_flat[test_indices], output_flat[test_indices]

train_x, train_y = torch.from_numpy(train_x).permute(2, 0, 1).unsqueeze(0).float(), torch.from_numpy(train_y).unsqueeze(0).unsqueeze(0).float()
val_x, val_y = torch.from_numpy(val_x).permute(2, 0, 1).unsqueeze(0).float(), torch.from_numpy(val_y).unsqueeze(0).unsqueeze(0).float()
test_x, test_y = torch.from_numpy(test_x).permute(2, 0, 1).unsqueeze(0).float(), torch.from_numpy(test_y).unsqueeze(0).unsqueeze(0).float()
print(train_x.shape)
print(train_y.shape)
print(train_x.dtype)
print(train_y.dtype)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from tqdm import tqdm

pos_weight = torch.tensor([30])

# def weighted_bce_loss(output, target, weights):
#     """
#     加权二元交叉矩阵损失。
#     output: 模型的输出
#     target: 目标标签
#     weights: 权重矩阵
#     """
#     loss = F.binary_cross_entropy_with_logits(output, target, reduction='none')
#     weighted_loss = loss * weights
#     return weighted_loss.mean()

# def extract_paths(segmentation):
#     import cv2

#     segmentation = segmentation.squeeze()

#     _, labels = cv2.connectedComponents(segmentation.astype(np.uint8))
#     return labels

# def path_continuity_loss(predicted, target, alpha=1):
#     predicted_paths = extract_paths(predicted.detach().cpu().numpy() > 0.5)
#     target_paths = extract_paths(target.detach().cpu().numpy() > 0.5)

#     loss = torch.abs(torch.tensor(predicted_paths).float() - torch.tensor(target_paths).float()).mean()

#     return alpha * loss

def connectivity_loss(mask, alpha = 0.001):
    kernel = torch.ones(3, 3).to(mask.device)
    dilated = F.conv2d(mask, kernel.unsqueeze(0).unsqueeze(0), padding=1)
    return torch.mean((dilated > 0).float() - mask) * alpha

class FocalLoss(nn.Module):
    def __init__(self, alpha=0.5, gamma=3):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        BCE_loss = F.binary_cross_entropy_with_logits(inputs, targets, reduction='none', pos_weight=pos_weight.to(device))
        targets = targets.type(torch.float32)
        at = self.alpha * targets + (1 - self.alpha) * (1 - targets)
        pt = torch.exp(-BCE_loss)
        F_loss = at * (1-pt)**self.gamma * BCE_loss
        return F_loss.mean()

class DoubleConv(nn.Module):
    """(convolution => [BN] => ReLU) * 2"""

    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.double_conv(x)

class Down(nn.Module):
    """Downscaling with maxpool then double conv"""

    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.maxpool_conv = nn.Sequential(
            nn.MaxPool2d(2),
            DoubleConv(in_channels, out_channels)
        )

    def forward(self, x):
        return self.maxpool_conv(x)

class Up(nn.Module):
    """Upscaling then double conv"""
    def __init__(self, in_channels, out_channels):
        super(Up, self).__init__()
        # 注意这里我们需要确保ConvTranspose2d的输入通道数等于上一层的输出通道数的两倍
        self.up = nn.ConvTranspose2d(in_channels // 2, in_channels // 2, kernel_size=2, stride=2)
        # 更新DoubleConv中输入通道数，因为它接收的是上采样的输出与跳跃连接的特征图的拼接
        self.conv = DoubleConv(in_channels, out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)
        # 输入是CHW
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]
        x1 = F.pad(x1, [diffX // 2, diffX - diffX // 2,
                        diffY // 2, diffY - diffY // 2])
        x = torch.cat([x2, x1], dim=1)
        return self.conv(x)

class OutConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(OutConv, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=1)

    def forward(self, x):
        return torch.sigmoid(self.conv(x))

class UNet(nn.Module):
    def __init__(self, n_channels, n_classes):
        super(UNet, self).__init__()
        self.n_channels = n_channels
        self.n_classes = n_classes

        self.inc = DoubleConv(n_channels, 64)
        self.down1 = Down(64, 128)
        self.down2 = Down(128, 256)
        self.down3 = Down(256, 512)
        self.down4 = Down(512, 512)
        self.up1 = Up(1024, 256)
        self.up2 = Up(512, 128)
        self.up3 = Up(256, 64)
        self.up4 = Up(128, 64)
        self.outc = OutConv(64, n_classes)

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)
        x = self.up1(x5, x4)
        x = self.up2(x, x3)
        x = self.up3(x, x2)
        x = self.up4(x, x1)
        logits = self.outc(x)
        return logits

# Initialize model, loss function, and optimizer
model = UNet(n_channels=3, n_classes=1).to(device)
criterion = FocalLoss()
# criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight.to(device))
optimizer = optim.AdamW(model.parameters(), lr=0.0005, weight_decay = 0.01)

training_loss_list = []
val_loss_list = []

# Training loop
def train():
    num_epochs = 200
    pbar = tqdm(range(num_epochs), desc="Loss: -- ; Progress")
    for epoch in pbar:
        model.train()
        train_loss = 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            train_labels = labels
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            continuity_loss = path_continuity_loss(outputs, labels)
            connectivity_loss_value = connectivity_loss(outputs)
            total_loss = loss + (connectivity_loss_value) * epoch / num_epochs
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * inputs.size(0)
        train_loss /= len(train_loader.dataset)

        # Validation loop
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                # print(labels)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)
        val_loss /= len(val_loader.dataset)

        if (epoch + 1) % 2 == 0:
            pbar.set_description(f"Loss: training {train_loss:.4f}, validation {val_loss:.4f}; Progress")
            # print(f'\nEpoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}\n')
        training_loss_list.append(train_loss)
        val_loss_list.append(val_loss)

train()
print("Training completed!")

iter = []
for i in range(len(val_loss_list)):
    iter.append(i + 1)

plt.plot(iter, training_loss_list, color = 'blue', label = 'training loss')
plt.plot(iter, val_loss_list, color = 'red', label = 'validation loss')
plt.ylim(0, 2)
plt.legend()
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

def tensor_to_numpy(outputs):
    # Assuming outputs is a batch of images, and we're interested in the first one
    # Also assuming binary classification with a sigmoid output where 0.5 is the threshold
    if torch.is_tensor(outputs):
        outputs = outputs.detach()  # detach from the graph
    outputs_np = outputs.cpu().numpy()  # Convert to numpy array
    predicted_image = outputs_np[0, 0]  # Get the first image, first channel
    predicted_image = (predicted_image > 0.5).astype(np.uint8)  # Threshold the output to get binary map
    return predicted_image

def evaluate_model(model, dataloader, device):
    model.eval()  # Set the model to evaluation mode
    correct = 0
    total = 0

    with torch.no_grad():  # No need to track gradients for evaluation
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)  # Get the index of the max log-probability
            # total += labels.size(0)
            total += 1200*2000
            correct += (predicted == labels).sum().item()

            predicted_image = tensor_to_numpy(outputs)

            # Plot the prediction
            plt.imshow(predicted_image, cmap='gray')
            plt.title('Predicted Road Map')
            plt.colorbar()
            plt.show()

            real_image = tensor_to_numpy(labels)

            # Plot the labels
            plt.imshow(real_image, cmap='gray')
            plt.title('Real Road Map')
            plt.colorbar()
            plt.show()

    accuracy = 100 * correct / total
    return accuracy

# Evaluate the model
accuracy = evaluate_model(model, test_loader, device)
print(f'Accuracy on the test set: {accuracy:.2f}%')

# Evaluate the model
accuracy = evaluate_model(model, val_loader, device)
print(f'Accuracy on the test set: {accuracy:.2f}%')

# Evaluate the model
accuracy = evaluate_model(model, train_loader, device)
print(f'Accuracy on the test set: {accuracy:.2f}%')
```
